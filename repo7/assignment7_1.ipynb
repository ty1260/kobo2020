まず, MeCabのサンプルプログラムを実行した.

```
import MeCab

m = MeCab.Tagger("-Ochasen")
m.parse('')
text  = '我輩は猫である。名前はまだ無い。'
node = m.parseToNode('我輩は猫である。名前はまだ無い。')
while node:
    # surfaceで単語(表層形)，featureで品詞情報がとれる
    print(node.surface, node.feature)
    node = node.next


m = MeCab.Tagger('-Owakati')
# parseはスペース区切りで返ってくるのでsplitでリストに．最後の単語は"\n"になっているので注意．
words = m.parse('我輩は猫である。名前はまだ無い。').split(" ")
print(words)
```

31. 動詞の原型をすべて抽出

neko.txtを一行ずつ読み込んで処理を行う. 当初, 品詞が動詞だった場合, node.surfaceを出力する形にしていたが, 重複があり全てが表示できなかったので, リストに追加して, 最後にset型に変換することで重複を取り除くことで解決した.

```
text_file = open('./neko.txt', 'r')
ans = []
for line in text_file:
  m = MeCab.Tagger("-Ochasen")
  m.parse('')
  node = m.parseToNode(line)
  while node:
    if node.feature.split(",")[0] == "動詞":
      ans.append(node.surface)
    node = node.next
text_file.close()
ans_unique = list(set(ans))
for i in ans_unique:
  print(i)
```

32. 動詞の原型をすべて抽出

```
text_file = open('./neko.txt', 'r')
ans = []
for line in text_file:
  m = MeCab.Tagger("-Ochasen")
  m.parse('')
  node = m.parseToNode(line)
  while node:
    if node.feature.split(",")[0] == "動詞":
      # print(node.feature.split(',')[6])
      ans.append(node.feature.split(',')[6])
    node = node.next
text_file.close()
ans_unique = list(set(ans))
for i in ans_unique:
  print(i)
```
